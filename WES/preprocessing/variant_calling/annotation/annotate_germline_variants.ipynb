```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

# LOAD RAW MAFS

```python
pd.set_option('display.max_columns', None)

folder = '/Users/jacobg/Dropbox/shared_leif/Jacob/WES_2024/germline/data/mafs'
files = sorted(os.listdir(folder))
germline_dfs = []
for f in files:
    print(f)
    if not f.endswith('.maf'):
        continue
    d = pd.read_csv(f'{folder}/{f}', header=1, sep='\t')
    #display(d)
    sample = f.replace('_HaplotypeCaller.maf', '')
    d['sample'] = sample
    germline_dfs.append(d)

germline = pd.concat(germline_dfs).dropna(axis=1, how='all')
display(germline)
```

# PREP DATA FOR ANNOVAR

```python
mandatory = ['Chromosome','Start_Position', 'End_Position','Reference_Allele', 'Tumor_Seq_Allele2']
optional = ['Tumor_Sample_Barcode', 't_depth', 't_ref_count', 't_alt_count', 'FILTER']
print(mandatory  + optional)
new = germline.loc[:, mandatory  + optional]
new
```

```python
path = '/Users/jacobg/Dropbox/shared_leif/Jacob/WES_2024/germline/data/combined/combined_germline_mutations_pre_annovar.csv' 
new.to_csv(path, sep = '\t', index=False, header=False)
```
