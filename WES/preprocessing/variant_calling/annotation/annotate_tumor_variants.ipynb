# ALL TUMOR VARIANTS

```python
pd.set_option('display.max_columns', None)

folder = '/Users/jacobg/Dropbox/shared_leif/Jacob/WES_2024/somatic/data/mutect2/filtered/maf'
files = os.listdir(folder)
cols = ['Tumor_Sample_Barcode', 'Chromosome', 'vcf_pos', 'Reference_Allele', 'Tumor_Seq_Allele2']
dfs = []
for f in files:
    print(f)
    if not f.endswith('.maf'):
        continue
    d = pd.read_csv(f'{folder}/{f}', header=1, sep='\t')
    sample = f.replace('_filtered.MuTect2.maf', '')
    d['sample'] = sample
    dfs.append(d)


df = pd.concat(dfs).dropna(axis=1, how='all')
display(df)
```
